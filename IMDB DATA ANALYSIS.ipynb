```python
country_films=imdb['country'].value_counts()[imdb['country'].value_counts()>=100 ]
print(country_films)
```

```python
plt.figure(figsize=(20, 8))
plt.bar(country_films.index, country_films.values)
plt.xlabel('Country')
plt.ylabel('Films')
plt.title('Number of films by Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
rating_votes=imdb[(imdb.imdbRating>8)]
print(rating_votes)
```

```python
top_rating=rating_votes['country'].value_counts()[rating_votes['country'].value_counts()>=10 ]
print(top_rating)
```

```python
plt.figure(figsize=(20, 8))
plt.bar(top_rating.index, top_rating.values)
plt.xlabel('Country')
plt.ylabel('Films')
plt.title('Number of Top-rated films by Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


genre_ratings = imdb[['genre', 'imdbRating']]


genre_counts = genre_ratings['genre'].value_counts()
filtered_genres = genre_counts[genre_counts > 500].index
filtered_genre_ratings = genre_ratings[genre_ratings['genre'].isin(filtered_genres)]

plt.figure(figsize=(10, 6))
plt.scatter(filtered_genre_ratings['genre'], filtered_genre_ratings['imdbRating'], alpha=0.5)
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.title('Individual Ratings for Genres with > 500 Films')
plt.xticks(rotation=45)
plt.show()
```

```python
genre_top_rating=rating_votes['genre'].value_counts()[rating_votes['genre'].value_counts()>=10 ]
print(genre_top_rating)
```

```python
plt.figure(figsize=(25, 10))
plt.bar(genre_top_rating.index, genre_top_rating.values)
plt.xlabel('Genre')
plt.ylabel('Films')
plt.title('Number of Top-rated films by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
genre_ratings = imdb.groupby('genre')['imdbRating'].mean()
genre_counts = imdb['genre'].value_counts()
filtered_genre_ratings = genre_ratings[genre_counts > 500]

print(filtered_genre_ratings)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(filtered_genre_ratings.index, filtered_genre_ratings.values)
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average Rating for Genres with > 500 Films')
plt.xticks(rotation=45)
plt.show()
```
